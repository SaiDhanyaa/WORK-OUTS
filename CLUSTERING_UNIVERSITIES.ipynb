```python
#IMPRT HIERARCHICAL CLUSTERING LIBRARIES
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
univ=pd.read_csv("/content/Universities.csv")
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(univ.iloc[:,1:])
```

## **SINGLE LINKAGE METHOD**

```python
# create dendrogram
dendrogram1 = sch.dendrogram(sch.linkage(df_norm, method='single'))
```

```python
# create clusters
hc1 = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'single')
```

```python
# save clusters for chart
y_hc1 = hc1.fit_predict(df_norm)
Clusters1=pd.DataFrame(y_hc1,columns=['Clusters'])
```

```python
Clusters1
```

```python
univ['h_clusterid1'] = Clusters1
```

```python
univ
```

```python
univ.iloc[:,1:].groupby(univ.h_clusterid1).mean()
```

## ***K-MEANS CLUSTERING***

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
univ = pd.read_csv("/content/Universities.csv")
```

```python
univ.head()
```

```python
# Normalization function 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_univ_df = scaler.fit_transform(univ.iloc[:,1:])
```

```python
# How to find optimum number of  cluster
#The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion:
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_univ_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# ***WE HAVE CHOSEN THE BEST K VALUE IS 4. AND WE ARE GOING WITH IT***

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(4, random_state=42)
clusters_new.fit(scaled_univ_df)
```

```python
clusters_new.labels_
```

# ***THESE ARE THE CLUSTER MEMBERS, THAT WE ARE GOING TO APPEND IT IN THE UNIVERSITY DATASET***

```python
#Assign clusters to the data set
univ['clust4'] = clusters_new.labels_
```

```python
univ
```

```python
univ.iloc[:,1:].groupby(univ.clust4).mean()
```

```python
univ 
```

## **INFERENCES:**

***The universities in the cluster 3 have higher SAT scores, higher in the top10, lowest student-facult ratio, highest exxpenses, and with highest grad rates among other clusters.So, the universiies in cluster 3 are the BEST UNIVERSITIES COMPARATIVELY ***
